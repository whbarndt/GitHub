# Load CSV - Add Price Column

```python
import numpy as np
import pandas as pd
from pathlib import Path
import yaml
```

```python
yaml_file_path = Path('config.local.finance.yml')
with open(yaml_file_path, 'r') as file:
    yaml_data = yaml.safe_load(file)
year = yaml_data['Year']
working_dir = yaml_data['Working_Dir']
```

```python
maui_expense_file = Path(f'Maui-Trip-2024.csv')
alcan_expense_file = Path(f'Alcan-Trip-2024.csv')
expense_df = pd.read_csv(working_dir / alcan_expense_file, comment='#')
date_col_name = 'Charge_Date'
expense_df[date_col_name] = pd.to_datetime(expense_df[date_col_name])
```

```python
price_col_name = 'Price'
total_expenses = expense_df[price_col_name].sum()
print(f'Total Expenses: ${total_expenses}')
```
